```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("DataSet/Train.csv")
```

```python
target = data.Col2
```

```python
DataDimension = 2395
```

```python
DataTenCol = data.iloc[:,1:DataDimension+1]
```

```python
colunsToBeDropped = []
for i, item in enumerate(DataTenCol.dtypes):
    if item.type not in [np.int64,np.float64]:
        colunsToBeDropped.append("Col"+str(i+2))
#         print(i+2)
#         print(item.type)
```

```python
colunsToBeDropped
```

```python
# len(DataTenCol.Col836.isnull().values)
```

```python
DataTenCol = DataTenCol.drop(colunsToBeDropped, axis=1)
```

```python
DataTenCol.head()
```

```python
input_dim = DataTenCol.shape[1]-1
```

```python
DataTenCol.head()
```

```python
DataTenCol.describe(include='all')
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='mean', missing_values = np.NaN)
```

```python
y = DataTenCol.iloc[:,0]
```

```python
X = DataTenCol.iloc[:,1:]
```

```python
X.isnull().values.any()
```

```python
X = pd.DataFrame(imp.fit_transform(X.values),index = X.index, columns = X.columns)
```

```python
X.isnull().values.any()
```

```python
# sns.pairplot(DataTenCol,hue='Col2')
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
updatedX = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
XTrain,XTest,YTrain,YTest = train_test_split(updatedX,y,test_size=0.001)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
import time
def base_model():
    model = Sequential()
    model.add(Dense(640, input_dim=input_dim, kernel_initializer='normal', activation='relu'))
    model.add(Dense(626, kernel_initializer='normal', activation='relu'))
    model.add(Dense(580, kernel_initializer='normal', activation='relu'))
    model.add(Dense(546, kernel_initializer='normal', activation='relu'))
    model.add(Dense(510, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(486, kernel_initializer='normal', activation='relu'))
    model.add(Flatten())
    model.add(Dropout(0.2))
    model.add(Dense(426, kernel_initializer='normal', activation='relu'))
    model.add(Dense(380, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(346, kernel_initializer='normal', activation='relu'))
    model.add(Dense(310, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(286, kernel_initializer='normal', activation='relu'))
    model.add(Flatten())
    model.add(Dropout(0.2))
    model.add(Dense(270, kernel_initializer='normal', activation='relu'))
    model.add(Dense(264, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(260, kernel_initializer='normal', activation='relu'))
    model.add(Dense(254, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(245, kernel_initializer='normal', activation='relu'))
    model.add(Dense(220, kernel_initializer='normal', activation='relu'))
    model.add(Flatten())
    model.add(Dropout(0.2))
    model.add(Dense(196, kernel_initializer='normal', activation='relu'))
    model.add(Dense(180, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(146, kernel_initializer='normal', activation='relu'))
    model.add(Dense(110, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(86, kernel_initializer='normal', activation='relu'))
    model.add(Flatten())
    model.add(Dropout(0.2))
    model.add(Dense(86, kernel_initializer='normal', activation='relu'))
    model.add(Dense(73, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(56, kernel_initializer='normal', activation='relu'))
    model.add(Dense(40, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(16, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal',activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
    return model
```

```python
model1 = base_model()
```

```python
model1.fit(XTrain,YTrain, batch_size=40,epochs=150)
```

```python
evaluation = model1.evaluate(XTrain,YTrain)
```

```python
evaluation = model1.evaluate(XTest,YTest)
```

```python
ypred = (model1.predict(XTest)>0.5)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ypred,YTest)*100
```
